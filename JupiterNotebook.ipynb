### Analyse tweets from Twitter with hashtags related to healthcare in the UK.

```python
# Set figure size.
sns.set(rc={'figure.figsize':(15, 12)})

# Set the plot style as white.
sns.set_style('white')

# Maximum column width to display
pd.options.display.max_colwidth = 200
```

```python
# Load the csv data using pd.read_csv.
tweets = pd.read_csv('tweets.csv')

# View the column names.
print(tweets.columns)
```

```python
# Explore the metadata.
print(tweets.info())
```

```python
# Determine whether there are missing values.
print(tweets.isnull().sum())
```

```python
# Explore the 'tweet retweet count'
retweet_counts = tweets['tweet_retweet_count'].value_counts()
print(retweet_counts)
```

```python
# Explore the 'tweet retweet count'
favorite_counts = tweets['tweet_favorite_count'].value_counts()
print(favorite_counts)
```

```python
# Create a new DataFrame containing only the text.
tweets_text = tweets[['tweet_full_text']].copy()

# View the DataFrame.
print(tweets_text)
```

```python
# Loop through the messages, and create a list of values containing the # symbol.
tags = []

for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
            
# Create a Pandas Series to count the values in the 'tags' list
tags_series = pd.Series(tags)
```

```python
# Display the first 30 records.
print(tags_series.head(30))
```

```python
# Calculate the count of each hashtag
hashtags_count = tags_series.value_counts()

# Convert the series to a DataFrame in preparation for visualisation.
tweetdata = hashtags_count.reset_index()

# Rename the columns.
tweetdata.columns = ['word', 'count']

# Fix the count datatype.
tweetdata['count'] = tweetdata['count'].astype(int)

# Display records where the count is larger than 10.
data_filtered = tweetdata['count'] > 10
result = tweetdata[data_filtered]

# View the result.
print(result)
```

```python
# Remove unrelated hashtags to healthcare in the UK hashtags
words_to_remove = ["#ai", "#job", "#strategy", "#marketing", "#tech", "#competitiveintelligence", "#meded",
                   "#hiring", "#news", "#machinelearning", "#technology",
                   "#competitivemarketing", "#foamed", "#education", "#innovation", "#usa"]

tweetdata.drop(tweetdata[tweetdata['word'].isin(words_to_remove)].index, inplace=True)

# Display records where the count is larger than 10.
data_filtered = tweetdata['count'] > 10
result = tweetdata[data_filtered]

# View the result.
print(result)
```

```python
# Create a Seaborn barplot displaying the remaining most frequently used hashtags (#)
sns.barplot(x='count', y='word', data=result)
plt.title("Top Trending Hashtags (#)")
plt.xlabel("Count")
plt.ylabel("Hashtags")
plt.show()

# Save the plot as a PNG file
plt.savefig('Top Trending Hashtags.png')
```

#### Question 1: Should the NHS start looking at increasing staff levels?

```python
# Create an aggregated data set to review the different features.
ar_agg = ar_filtered.groupby(['appointment_month', 'hcp_type', 'appointment_status', 'appointment_mode', 'time_between_book_and_appointment'])['count_of_appointments'].sum().reset_index()

# View the DataFrame.
print(ar_agg)
```

```python
# Determine the total number of appointments per month.
ar_df = ar_filtered.groupby('appointment_month')['count_of_appointments'].sum().reset_index()

# Add a new column to indicate the average utilisation of services.
# Monthly aggregate / 30 to get to a daily value.
ar_df['utilisation'] = ar_df['count_of_appointments'] / 30
ar_df['utilisation'] = ar_df['utilisation'].apply(lambda x: min(x, 1200000))
ar_df['utilisation'] = ar_df['utilisation'].round(1)

# View the DataFrame.
print(ar_df)
```

```python
# Plot sum of count of monthly visits.
# Convert the appointment_month to string data type for ease of visualisation.
ar_agg['appointment_month'] = ar_agg['appointment_month'].astype(str)
ar_df['appointment_month'] = ar_df['appointment_month'].astype(str)
```

```python
# Create a lineplot with Seaborn.
plt.figure(figsize=(12, 6)) 
sns.lineplot(data=ar_agg, x='appointment_month', y='count_of_appointments', marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Monthly Visits')
plt.title('Number of Monthly Visits Over Time')
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()  
plt.show()

# Save the plot as a PNG file
plt.savefig('Number of Monthly Visits Over Time.png')
```

```python
# Plot monthly capacity utilisation.
# Create a lineplot.
plt.figure(figsize=(12, 6))
sns.lineplot(data=ar_df, x='appointment_month', y='utilisation',)
plt.xlabel('Month')
plt.ylabel('Monthly Capacity Utilisation')
plt.title('Monthly Capacity Utilisation Over Time')

# Set a horizontal line at the maximum capacity (1,200,000) for reference
plt.axhline(y=1200000, color='red', linestyle='--', label='Maximum Capacity')
plt.xticks(rotation=45)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()

# Save the plot as a PNG file
plt.savefig('Monthly Capacity Utilisation.png')
```
